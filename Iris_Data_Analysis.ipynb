```python
#1. kütüphanelerin içe aktarılması
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#2. Veri dosyasının yüklenmesi
df=pd.read_csv('/content/drive/MyDrive/BTK-3-YZ/data/iris.csv')
```

```python
print(df)
```

```python
df.info()
```

```python
print(df.groupby('Species').count()) 
```

```python
df.groupby("Species").sum()
```

```python
df.groupby("Species").mean()
```

```python
df=df.drop("Id",axis=1)
df.groupby("Species").agg([min,max,np.std])
```

```python
sns.scatterplot(x="Species",y="SepalLengthCm",data=df)
plt.title("SepalLengthCm")
plt.show()
```

```python
sns.scatterplot(x="SepalWidthCm",y="SepalLengthCm",hue='Species', data=df)
plt.title("SepalLengthCm-Sepal WidthCm")
plt.show()
```

```python
sns.scatterplot(x="PetalLengthCm",y="SepalLengthCm",hue='Species', data=df)
plt.title("SepalLengthCm-Sepal WidthCm")
plt.show()
```

```python
sns.scatterplot(x="PetalWidthCm",y="SepalLengthCm",hue='Species', data=df)
plt.title("SepalLengthCm-Sepal WidthCm")
plt.show()
```
